# SVM

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
```

### 数据加载

```python
f = open("twofeature.txt",'r+')
data = f.readlines()
data = [i.split() for i in data]
y = [int(i[0]) for i in data]
x = [[float(j[2:]) for j in i[1:]] for i in data ]
```

```python
x = np.array(x)
y = np.array(y)
l_x = np.where(y == 1)
o_x = np.where(y == -1)
plt.plot(x[l_x][:,0],x[l_x][:,1],'o',label = 'label = 1')
plt.plot(x[o_x][:,0],x[o_x][:,1],'o',label = 'label = 2')
plt.legend()
```

```python
def plot_svc_decision_function(clf, ax=None):
    """Plot the decision function for a 2D SVC"""
    if ax is None:
        ax = plt.gca()
    x = np.linspace(plt.xlim()[0], plt.xlim()[1], 30)
    y = np.linspace(plt.ylim()[0], plt.ylim()[1], 30)
    
    Y, X = np.meshgrid(y, x)
    P = np.zeros_like(X)
    for i, xi in enumerate(x):
        for j, yj in enumerate(y):
            P[i, j] = clf.decision_function(np.array([xi,yj]).reshape(1,-1))
    # plot the margins
    ax.contour(X, Y, P, colors='k',
               levels=[-1, 0, 1], alpha=0.5,
               linestyles=['--', '-', '--'])
```

```python
clf = svm.SVC(C=1, kernel='linear', gamma=20, decision_function_shape='ovr')
clf.fit(x, y)
plt.scatter(x[:, 0], x[:, 1], c=y, s=50, cmap='spring')
plot_svc_decision_function(clf)
plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1],
            s=200, facecolors='none');
plt.xlabel('c = 1 train score = ' + str(clf.score(x, y)))
```

```python
clf = svm.SVC(C=100, kernel='linear', gamma=20, decision_function_shape='ovr')
clf.fit(x, y)
plt.scatter(x[:, 0], x[:, 1], c=y, s=50, cmap='spring')
plot_svc_decision_function(clf)
plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1],
            s=200, facecolors='none');
plt.xlabel('c = 1 train score = ' + str(clf.score(x, y)))
```

### 电子邮件分类

### train ：50

```python
f = open("email_train-50.txt",'r+')
data = f.readlines()
data = [i.split() for i in data]
y_train = [int(i[0]) for i in data]
x_train= [[j.split(':') for j in i[1:]] for i in data ]
x_train = [[[int(k) for k in j]for j in i] for i in x_train]
x_new_train = []
for i in x_train:
    x_array = np.zeros(3000)
    for j in i:
        x_array[j[0]] = j[1]
    x_new_train.append(x_array)
x_train = np.array(x_new_train)
y_train = np.array(y_train)
```

```python
f = open("email_test.txt",'r+')
data = f.readlines()
data = [i.split() for i in data]
y_test = [int(i[0]) for i in data]
x_test= [[j.split(':') for j in i[1:]] for i in data ]
x_test = [[[int(k) for k in j]for j in i] for i in x_test]
x_new_test = []
for i in x_test:
    x_array = np.zeros(3000)
    for j in i:
        x_array[j[0]] = j[1]
    x_new_test.append(x_array)
x_test = np.array(x_new_test)
y_test = np.array(y_test)
```

### 准确率

```python
cl = svm.SVC(C=100, kernel='linear', gamma=2000, decision_function_shape='ovr')
cl.fit(x_train, y_train)
cl.score(x_test,y_test)
```
